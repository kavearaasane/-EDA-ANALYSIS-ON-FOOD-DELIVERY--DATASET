```python
# visuialize location as per order
datamap={
'Banashankari':3,
'Basavanagudi':2,
'Whitefield':1,
'Jayanagar':4,
'Vijayroad':5,
'Kumaraswamy Layout':6,
'Rajarajeshwari Nagar':7,
'Kumaraswamy Layout':8,
'Mysore Road':9,
'JP Nagar':10,
'Uttarahalli':11,
'Uttarahalli':12
}
df['location']=df['location'].map(datamap)
datmap={
    'Yes':1,
    'No':0
}
df['online_order']=df['online_order'].map(datmap)
import matplotlib.pyplot as plt
x=df['location']
y=df['online_order']
plt.xlabel('location')
plt.ylabel('online_order')
plt.scatter(x,y,color='green')
plt.show()
```

```python
# Visualize Restaurant VS Rate.
location_plt=pd.crosstab(df['rate'],df['rest_type'])
location_plt.plot(kind='bar',stacked=True);
plt.title('Restaurant type vs Rating',fontsize=9,fontweight='normal')
plt.ylabel('Restaurant type',fontsize=12,fontweight='normal')
plt.xlabel('Rate',fontsize=11,fontweight='normal')
plt.xticks(fontsize=9,fontweight='normal')
plt.yticks(fontsize=9,fontweight='normal');
plt.legend().remove();
```

```python
#ouliers 
import pandas as pd
df=pd.read_csv('zomato.csv')
import seaborn as sns
sns.boxplot(data=df,x=df['votes'])
Q1=df['votes'].quantile(0.25)
Q3=df['votes'].quantile(0.75)
IQR=Q3-Q1
print('Q1:' ,Q1)
print('Q3:' ,Q3)
print('IQR:',Q3)
minimum_value = Q1-1.5*IQR
maximum_value = Q3+1.5*IQR
print('the outliers is :')
print(minimum_value,maximum_value)
```
